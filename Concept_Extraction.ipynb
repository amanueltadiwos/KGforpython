# Concept Mapping based on Code **elements**

```
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

**Read the file**

```
import io
df = pd.read_csv(io.StringIO(uploaded["Labeled_data_terms.csv"].decode("utf-8")), sep=',')
data = df[['word','label']]
data.head(10)
```

```
agg_func = lambda s: [(w, t) for w,t in zip(s["word"].values.tolist(),
                                                        s["label"].values.tolist())]
concepts = data.groupby("label").apply(agg_func)
for item in concepts:
        print(item)
```

```
import csv
with open('ConceptsExtracted.csv', mode='w') as cs:
  for items in concepts:
    for i in items:
      concept_writer = csv.writer(cs, delimiter=',')
      concept_writer.writerow([i])
```
